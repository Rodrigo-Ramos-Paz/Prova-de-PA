```python
#Pedir um número ao usuário e fazer a soma dos números até o número dado

a = int(input("Digite um número: "))
soma_total = 0
for i in range(1,a +1):
  soma_total += i
print(f"A soma dos números até {a} é igual {soma_total}")
```

```python
#Solicite um número ao usuário e mostre sua tabuada inteira

num_tabuada = int(input("Digite um número: "))
for i in range (1,11):
  print(f"{num_tabuada} x {i} = {num_tabuada * i} ")
```

```python
#Pedir um número inteiro positivo ao usuário e calcular o seu fatorial

b = int(input("Digite um número inteiro e positivo: "))
fatorial = 1
if (b) >= 0:
  for i in range(1,b+1):
   fatorial *= i
print(f"O fatorial de {b} é: {fatorial}")
```
